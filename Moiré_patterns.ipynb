# Moiré Patterns

patterns emerging from overlapping grids

```python
from matplotlib import pyplot as plt
import math
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
plt.figure(figsize = (12, 12))

def plot(x,y): plt.plot(x,y, color="black")

def grid(
    x0  = 0.,
    y0  = 0.,
    dx  = 1e-2, 
    dy  = 1e-2, 
    phi = 0,
    ):
  
  # optimize the angle (in radians)
  phi %= np.pi
  tmp  = phi - np.pi/2
  phi  = tmp if abs(tmp) < phi else phi

  # tilt grid constants
  dx /= np.cos(phi)
  dy /= np.cos(phi)
  x0 /= np.cos(phi)
  y0 /= np.cos(phi)
  
  
  # more horizontal lines:
  x = np.linspace(-1, 2, 2)
  for i in range(0, math.ceil(3 / dy)):
    y = np.tan(phi) * x + dy * i + (y0 - 1)
    plot(x,y)

  # more vertical lines:
  y = x # inverse function
  for i in range(0, math.ceil(3 / dx)):
    x = -np.tan(phi) * y + dx * i + (x0 - 1)
    plot(x,y)


grid(dx = .008, dy = .008, phi = np.pi/180 * 0)
grid(dx = .008, dy = .008, phi = np.pi/180 * 2)

plt.xlim((0, 1))
plt.ylim((0, 1))
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax  = plt.axes()

def tick(i):
  ax.set_xlim((0,1))
  ax.set_ylim((0,1))
  grid(dx = .007, dy = .007, phi = np.pi/180 * 0)
  grid(dx = .007, dy = .007, phi = np.pi/180 * i/10)
  print("Progress: {} ".format(i))
  return ax.plot()

ani = FuncAnimation(
    fig = fig,
    func = tick,
    frames = 50,
    interval = 70,
    blit = True,
)

ani.save("moire.gif", writer="imagemagick")
```
